```python
from PIL import Image, ImageTk
import cv2 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from tensorflow.keras.models import load_model
import numpy as np
from tensorflow.keras.utils import load_img, img_to_array
from keras.models import load_model
font=cv2.FONT_HERSHEY_COMPLEX

model = load_model('traffic_sign.h5')

video=cv2.VideoCapture(0)


while True:
    ret,frame=video.read()
    crop_img=frame[100:300,100:300]
    name='E:/Anh/'+'test.jpg'
    cv2.imwrite(name, crop_img)
    img = load_img(name,target_size=(150,150))
    img = img_to_array(img)
    img = img.reshape(1,150,150,3)
    img = img.astype('float32')
    img = img/255
    kq= np.argmax(model.predict(img),axis=-1)
    cv2.rectangle(frame, (100,100), (300, 300), (0,255,0), 3)
    if(kq==0):
        cv2.putText(frame, 'bien bao cam',(100+20,300+28), font, 0.75, (255,0,255),1, cv2.LINE_AA)
    if(kq==1):
        cv2.putText(frame, 'cam nguoc chieu',(100+20,300+28), font, 0.75, (255,0,255),1, cv2.LINE_AA)
    if(kq==2):
        cv2.putText(frame, 'cam vuot',(100+20,300+28), font, 0.75, (255,2,255),1, cv2.LINE_AA)
    if(kq==3):
        cv2.putText(frame, 'cam xe tai vuot',(100+20,300+28), font,  0.75, (255,2,255),1, cv2.LINE_AA)
    if(kq==4):
        cv2.putText(frame, 'duong hep ben phai',(100+20,300+28), font,  0.75, (255,2,255),1, cv2.LINE_AA)
    if(kq==5):
        cv2.putText(frame, 'duong uu tien',(100+20,300+28), font,  0.75, (255,2,255),1, cv2.LINE_AA)
    if(kq==6):
        cv2.putText(frame, 'nguy hiem khac',(100+20,300+28), font,  0.75, (255,2,255),1, cv2.LINE_AA)
    if(kq==7):
        cv2.putText(frame, 'noi giao nhau chay theo vong xuyen',(100+20,300+28), font,  0.75, (255,2,255),1, cv2.LINE_AA)
    if(kq==8):
        cv2.putText(frame, 'Stop',(100+20,300+28), font,  0.75, (255,2,255),1, cv2.LINE_AA)
    if(kq==9):
        cv2.putText(frame, 'toc do toi da 60',(100+20,300+28), font,  0.75, (255,2,255),1, cv2.LINE_AA)
    cv2.imshow("Detect Traffic sign", frame)
    k=cv2.waitKey(1)
    if k==ord('v') :
        break
video.release()
cv2.destroyAllWindows()
```
