```python
# load libraries
import tkinter as tk
from tkinter import filedialog
from tkinter import *
from PIL import ImageTk, Image
import numpy
from tensorflow.keras.utils import load_img, img_to_array
import numpy as np
#load the trained model to detect 10 traffic signs
from keras.models import load_model
model = load_model('traffic_sign.h5')
#Classes Label
classes = { 
    0:'bien_duong_cam',
    1:'cam nguoc chieu',
    2:'cam vuot',
    3:'cam xe tai vuot',
    4:'duong hep ben phai',
    5:'duong uu tien',
    6:'nguy hiem khac',
    7:'noi giao nhau chay theo vong xuyen',
    8:'stop',
    9:'toc_do_toida_60',
}
#create Gui
top=tk.Tk()
top.geometry('800x700')
top.title('Nhận Diện 10 Biển Báo')
top.configure(background='#FFDF00')
label=Label(top,background='#CDCDCD', font=('Times New Roman',15,'bold'))
sign_image = Label(top)
def classify(file_path):
    global label_packed    
    img = load_img(file_path,target_size=(150,150))
    img = img_to_array(img)
    img = img.reshape(1,150,150,3)
    img = img.astype('float32')
    img = img/255
    kq= np.argmax(model.predict(img),axis=-1)
    
    if(kq==0):
        kq = "biển đường cấm: cấm tất cả các loại phương tiện đi lại cả hai hướng,\n trừ các xe được ưu tiên theo quy định. "
    if(kq==1):
        kq = "cấm ngược chiều: cấm tất cả các loại xe đi vào theo chiều đặt biển,\n trừ các xe được ưu tiên theo quy định "
    if(kq==2):
        kq = "cấm vượt: cấm các loại xe cơ giới vượt nhau kể cả xe được ưu tiên theo quy định."
    if(kq==3):
        kq = "cấm xe tải vượt: cấm các loại ôtô tải vượt xe cơ giới khác.\n Biển có hiệu lực cấm tất cả các loại ôtô tải có trọng lượng lớn nhất cho phép (bao gồm trọng lượng xe và hàng)."
    if(kq==4):
        kq = "đường hẹp bên phải: Báo trước sắp đến một đoạn đường bị hẹp đột ngột về phía phải."
    if(kq==5):
        kq = "đường ưu tiên: bắt đầu đoạn đường ưu tiên"
    if(kq==6):
        kq = "Nguy hiểm khác: báo trên đường có nguy hiểm"
    if(kq==7):
        kq = "Nơi giao nhau chạy theo vòng xuyến: cấm các loại\n xe cơ giới vượt nhau kể cả xe được ưu tiên theo quy định."
    if(kq==8):
        kq = "Stop: Biển báo giao thông có hiệu lực buộc các xe cơ giới\n và thô sơ kể cả các xe được ưu tiên theo luật lệ Nhà nước quy định dừng lại trước biển hoặc trước vạch ngang đường."
    if(kq==9):
        kq = "tốc độ tối đa 60 Km/h: Tốc độ tối đa cho phép"
    
    label.configure(foreground='#da2616', text=kq) 
    
def show_classify_button(file_path):
    classify_b=Button(top,text="Nhận Diện",
   command=lambda: classify(file_path),padx=10,pady=15)
    classify_b.configure(background='#DC143C', foreground='white',
font=('Times New Roman',13,'bold')) #recognize
    classify_b.place(relx=0.69,rely=0.46)
def upload_image():
    try:
        file_path=filedialog.askopenfilename()
        uploaded=Image.open(file_path)
        uploaded.thumbnail(((top.winfo_width()/2.25),
    (top.winfo_height()/2.25)))
        im=ImageTk.PhotoImage(uploaded)
        sign_image.configure(image=im)
        sign_image.image=im
        label.configure(text='')
        show_classify_button(file_path)
    except:
        pass
upload=Button(top,text="Tải Ảnh Lên",command=upload_image,
  padx=10,pady=15)
upload.configure(background='#32CD32', foreground='white',
    font=('Times New Roman',13,'bold')) #upload
upload.pack(side=BOTTOM,pady=50)
sign_image.pack(side=BOTTOM,expand=True)
label.pack(side=BOTTOM,expand=True)
heading = Label(top, text="Nhận Diện Biển Báo ",pady=15, font=('Times New Roman',20,'bold'))
heading.configure(background='#32CD32',foreground='white')
heading.pack()
top.mainloop()
```
